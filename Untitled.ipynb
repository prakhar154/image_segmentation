```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

reading sample data

```python
img = cv2.imread("sample_image.jpg")
```

```python
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
x = img.reshape((-1,3))
```

making a cluster for image colors

```python
km = KMeans(n_clusters=5)
km.fit(x)
```

```python
center = km.cluster_centers_
```

```python
colors = []
for i,color in enumerate(center):
    plt.subplot(1, 10, i+1)
    colors.append(color)
    
    square = np.zeros((100, 100, 3), dtype = 'uint8')
    square[:, :, :] = color
    plt.imshow(square)
    plt.axis("off")
plt.show()
```

```python
np.unique(km.labels_, return_counts=True)
```

```python
labels = km.labels_
```

```python
new_img = np.zeros(x.shape)
for i in range(x.shape[0]):
    new_img[i] = colors[labels[i]]
print(new_img.shape)    
```

```python
new_img = new_img.reshape(img.shape)
print(img.shape)
print(new_img.shape)
```

```python
new_img /=255
plt.imshow(new_img)
plt.axis("off")
plt.title("Segmented image ")
plt.show()
```
